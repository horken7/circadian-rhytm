```python
# magnitude of acceleration
magnitude = df['wearable-mag-xl1']
datetime = pd.to_datetime(df.index)

fig = plt.figure(figsize=(15,10))
ax = fig.add_subplot(111)
ax.set_ylabel('Magnitude of acceleration', fontsize=18)
ax.set_xlabel('Time', fontsize=18)
ax.plot(datetime, magnitude)
```

```python
# raw accelerometer data
accelerometer_directions = df[['wearable-xl1-x', 'wearable-xl1-y', 'wearable-xl1-z']]

# get rid of rows with missing values
accelerometer_directions = accelerometer_directions.dropna()
datetime = pd.to_datetime(accelerometer_directions.index)

# arm angle calculation using accelerometer data
accelerometer_directions['angle'] = 180/np.pi * np.arctan(accelerometer_directions['wearable-xl1-z'] / np.sqrt( np.square(accelerometer_directions['wearable-xl1-x']) + np.square(accelerometer_directions['wearable-xl1-z']) ))

fig = plt.figure(figsize=(15,10))
ax = fig.add_subplot(111)
ax.set_ylabel('arm angle [degrees]', fontsize=18)
ax.set_xlabel('Time', fontsize=18)
ax.plot(datetime, accelerometer_directions['angle'])
```

model for arm angle:  $$ \theta_{\text{arm}} = tan^{-1} \bigg(\frac{ a_{z} }{ \sqrt{a_{x}^{2} + a_{y}^{2}} } \bigg) $$

```python
# convert index to DatetimeIndex
accelerometer_directions.index = pd.DatetimeIndex(accelerometer_directions.index)

# resample with 5 min interval
resample_5min_accel = accelerometer_directions.resample('5Min')


fig = plt.figure(figsize=(15,10))
ax = fig.add_subplot(111)
ax.set_title('Arm angle std with 5 min resampling')
ax.set_ylabel('std')
ax.set_ylabel('time')
ax.plot(resample_5min_accel.index, resample_5min_accel.std()['angle'])
```

```python
def classify(val):
    zone0 = 3
    zone1 = 6
    zone2 = 10
    zone3 = 20
    if(val <= zone0):
        return 0
    elif(val > zone0 and val <= zone1):
        return 1
    elif(val > zone1 and val <= zone2):
        return 2 
    elif(val > zone2 and val <= zone3):
        return 3
    else:
        return 4
    
ind = resample_5min_accel.index
std = accelerometer_directions.resample('5Min').std()['angle']
classification = []
for i, val in enumerate(std.values):
    classification.append(classify(val))
    
print('Time in Zone0: '+str((classification.count(0)*5)/60)+'h')
print('Time in Zone1: '+str((classification.count(1)*5)/60)+'h')
print('Time in Zone2: '+str((classification.count(2)*5)/60)+'h')
print('Time in Zone3: '+str((classification.count(3)*5)/60)+'h')
print('Time in Zone4: '+str((classification.count(4)*5)/60)+'h')
print('Total time in bed: '+str(len(classification)*5/60)+'h')

print(classification.count(4)/len(classification))
    
plt.figure(1)
fig = plt.figure(figsize=(15,10))
ax = fig.add_subplot(111)
ax.set_title('Classified arm angle std')
ax.set_xlabel('Time', fontsize=18)
ax.set_ylabel('Arm angle std', fontsize=18)
colors = ['r','g','b','y','c']
for i, n in enumerate(std):
    plt.plot(ind[i], std[i], 'o', c=colors[classification[i]])
import matplotlib.patches as mpatches
import matplotlib.pyplot as plt

red_patch = mpatches.Patch(color='red', label='Zone1: std <= 3')
green_patch = mpatches.Patch(color='green', label='Zone2: 3 < std <= 6')
blue_patch = mpatches.Patch(color='blue', label='Zone3: 6 < std <= 10')
yellow_patch = mpatches.Patch(color='yellow', label='Zone4: 10 < std <= 20')
cyan_patch = mpatches.Patch(color='cyan', label='Zone5: std > 20')
plt.legend(handles=[red_patch, green_patch, blue_patch, yellow_patch ,cyan_patch])
plt.legend()
plt.show()
```

```python
deviation_5min = resample_5min_accel['angle'].max()-resample_5min_accel['angle'].min()
mean_5min = resample_5min_accel.mean()

fig = plt.figure(figsize=(15,10))
ax = fig.add_subplot(111)
ax.set_title('Arm angle with 5 min resampling')
ax.set_ylabel('arm angle')
ax.plot(resample_5min_accel.index, deviation_5min)
ax.plot(resample_5min_accel.index, mean_5min['angle'])
lines, labels = ax.get_legend_handles_labels()
ax.legend(lines[:], ['Maximum movement', 'Mean angle'], loc='best') 


```

```python
classification = []
for ind, val in enumerate(deviation_5min.values):
    classification.append(classify(val))

plt.figure(1)
fig = plt.figure(figsize=(15,10))
ax = fig.add_subplot(111)
ax.set_title('Classified arm angle deviation over 5min intervals')
ax.set_ylabel('arm angle')
ax.set_ylabel('time')
colors = ['r','g','b','y','c']
ind = deviation_5min.index
for i, n in enumerate(deviation_5min):
    plt.plot(ind[i], n, 'o', c=colors[classification[i]])
red_patch = mpatches.Patch(color='red', label='Zone1: angle <= 3')
green_patch = mpatches.Patch(color='green', label='Zone2: 3 < angle <= 6')
blue_patch = mpatches.Patch(color='blue', label='Zone3: 6 < angle <= 10')
yellow_patch = mpatches.Patch(color='yellow', label='Zone4: 10 < angle <= 20')
cyan_patch = mpatches.Patch(color='cyan', label='Zone5: angle > 20')
plt.legend(handles=[red_patch, green_patch, blue_patch, yellow_patch ,cyan_patch])
plt.legend()
```
