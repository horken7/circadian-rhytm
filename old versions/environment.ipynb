```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import os

%matplotlib inline
```

```python
# search for csv location

found = []

def find(name, path):
    path = os.getcwd() + path
    file = name

    for root, dir, files in os.walk(path):
        for item in files:
            if file == item:
                return os.path.join(root, file)
        return None
data = ['00', '01', '02', '03', '05', '22', '23', '24', '25']
for d in data:
    found.append(find('data_' + d + '.csv.gz', '/irc-sphere-sleep-56db93f64661/sphere-sensor-data'))
    found.append(find('data_' + d + '.csv', '/irc-sphere-sleep-56db93f64661/sphere-sensor-data'))
```

```python
# load file

df = []
for path in found:
    if path is not None:
        if 'gz' in path:
            fileName = path
            df.append(pd.read_csv(fileName, index_col='datetime', compression='gzip'))
        else: 
            fileName = path
            df.append(pd.read_csv(fileName, index_col='datetime'))
```

```python
environment_data = []
for d in df:
    environment_data.append(d[['cold-water', 'door', 'electricity',
                           'hid', 'hot-water', 'humidity',
                           'light', 'motion', 'noise', 'temperature']])
environment_data[0]
```

```python
fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(111)

for e in environment_data:
    # exclude missing data
    feature = e['temperature'].dropna()

    resample_interval = '5Min'

    # convert index to DatetimeIndex
    feature.index = pd.DatetimeIndex(feature.index)

    feature_resample = feature.resample(resample_interval).mean()
    
    timeindex = []
    for i, t in enumerate(feature_resample.index):
        timeindex.append(i)

    plt.plot(timeindex, feature_resample)
```
